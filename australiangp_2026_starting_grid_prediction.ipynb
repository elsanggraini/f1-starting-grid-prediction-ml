```python
import os
import fastf1
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# bikin folder cache dulu
os.makedirs('/content/f1_cache', exist_ok=True)

fastf1.Cache.enable_cache('/content/f1_cache')
```

```python
years = [2019, 2022, 2023, 2024]
all_data = []

for year in years:
    session = fastf1.get_session(year, 'Australia', 'Q')
    session.load()

    quali = session.results[
        ['Abbreviation', 'FullName', 'TeamName', 'Position']
    ]
    quali['Year'] = year
    all_data.append(quali)

df = pd.concat(all_data).dropna()
df.head()
```

```python
# Rata-rata posisi qualifying per driver (Australia GP)
driver_avg = (
    df.groupby('Abbreviation')['Position']
    .mean()
    .reset_index()
)
driver_avg.columns = ['DriverCode', 'AvgDriverQuali']

# Rata-rata posisi qualifying per tim
team_avg = (
    df.groupby('TeamName')['Position']
    .mean()
    .reset_index()
)
team_avg.columns = ['Team', 'AvgTeamQuali']

# Dataset training
data = df.merge(driver_avg, left_on='Abbreviation', right_on='DriverCode')
data = data.merge(team_avg, left_on='TeamName', right_on='Team')

X = data[['AvgDriverQuali', 'AvgTeamQuali']]
y = data['Position']
```

```python
model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

model.fit(X, y)

# Evaluasi di data training (baseline)
y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"MAE  : {mae:.2f}")
print(f"R   : {r2:.3f}")
```

```python
drivers_2026 = [
    {'DriverCode': 'VER', 'Team': 'Red Bull Racing'},
    {'DriverCode': 'HAD', 'Team': 'Red Bull Racing'},
    {'DriverCode': 'LEC', 'Team': 'Ferrari'},
    {'DriverCode': 'HAM', 'Team': 'Ferrari'},
    {'DriverCode': 'NOR', 'Team': 'McLaren'},
    {'DriverCode': 'PIA', 'Team': 'McLaren'},
    {'DriverCode': 'ANT', 'Team': 'Mercedes'},
    {'DriverCode': 'RUS', 'Team': 'Mercedes'},
    {'DriverCode': 'ALB', 'Team': 'Williams'},
    {'DriverCode': 'SAI', 'Team': 'Williams'},
    {'DriverCode': 'LAW', 'Team': 'RB'},
    {'DriverCode': 'LID', 'Team': 'RB'},
    {'DriverCode': 'BOR', 'Team': 'Audi'},
    {'DriverCode': 'HUL', 'Team': 'Audi'},
    {'DriverCode': 'ALO', 'Team': 'Aston Martin'},
    {'DriverCode': 'STR', 'Team': 'Aston Martin'},
    {'DriverCode': 'GAS', 'Team': 'Alpine'},
    {'DriverCode': 'COL', 'Team': 'Alpine'},
    {'DriverCode': 'BOT', 'Team': 'Cadillac'},
    {'DriverCode': 'PER', 'Team': 'Cadillac'},
    {'DriverCode': 'BEA', 'Team': 'Haas F1 Team'},
    {'DriverCode': 'OCO', 'Team': 'Haas F1 Team'},
]

grid_2026 = pd.DataFrame(drivers_2026)
grid_2026
```

```python
# =========================
# CELL 7  DATA FUSION (FIXED)
# =========================

# 1. Grid Average (Australia Quali historis)
grid_average = df['Position'].mean()
print(f"Grid Average Qualifying Position: {grid_average:.2f}")

# 2. Merge driver average
grid_2026 = grid_2026.merge(
    driver_avg,
    on='DriverCode',
    how='left'
)

# 3. Merge team average
grid_2026 = grid_2026.merge(
    team_avg,
    on='Team',
    how='left'
)

# 4. Rename kolom biar konsisten
grid_2026 = grid_2026.rename(columns={
    'Position_x': 'AvgDriverQuali',
    'Position_y': 'AvgTeamQuali'
})

# 5. Fallback logic (GRID AVERAGE MODE)
grid_2026['AvgDriverQuali'] = grid_2026['AvgDriverQuali'].fillna(
    grid_2026['AvgTeamQuali']
)

grid_2026['AvgDriverQuali'] = grid_2026['AvgDriverQuali'].fillna(
    grid_average
)

grid_2026['AvgTeamQuali'] = grid_2026['AvgTeamQuali'].fillna(
    grid_average
)

# 6. Final check
grid_2026
```

```python
# =========================
# CELL 7.5  DRIVER CODE MAPPING
# =========================

driver_code_map = {
    'Max Verstappen': 'VER',
    'Sergio Perez': 'PER',
    'Charles Leclerc': 'LEC',
    'Carlos Sainz': 'SAI',
    'Lewis Hamilton': 'HAM',
    'George Russell': 'RUS',
    'Lando Norris': 'NOR',
    'Oscar Piastri': 'PIA',
    'Fernando Alonso': 'ALO',
    'Lance Stroll': 'STR',
    'Esteban Ocon': 'OCO',
    'Pierre Gasly': 'GAS',
    'Valtteri Bottas': 'BOT',
    'Zhou Guanyu': 'ZHO',
    'Kevin Magnussen': 'MAG',
    'Nico Hulkenberg': 'HUL',
    'Yuki Tsunoda': 'TSU',
    'Daniel Ricciardo': 'RIC',
    'Alexander Albon': 'ALB',
    'Liam Lawson': 'LAW',
    'Isack Hadjar': 'HAD',
    'Arvid Linblad': 'LID',
    'Kimi Antonelli': 'ANT',
    'Oliver Bearman': 'BEA',
    'Gabriel Bortoleto': 'BOR',
    'Franco Colapinto': 'COL',
}

# Tambahkan DriverCode ke data historis
df['DriverCode'] = df['FullName'].map(driver_code_map)

# Buang driver yang ga kepake / ga kemapping
df = df.dropna(subset=['DriverCode'])

df.head()
```

```python
# =========================
# CELL 7.6  STANDARDIZE TEAM NAME
# =========================

# Samakan nama kolom tim
df = df.rename(columns={'TeamName': 'Team'})

df.columns
```

```python
# =========================
# CELL 8  PREPARE TRAINING DATA
# =========================

# Gabungkan avg driver & avg team ke data historis
df_model = df.merge(
    driver_avg.rename(columns={'Position': 'AvgDriverQuali'}),
    on='DriverCode',
    how='left'
)

df_model = df_model.merge(
    team_avg.rename(columns={'Position': 'AvgTeamQuali'}),
    on='Team',
    how='left'
)

# Fallback (kalau ada yang bolong)
df_model['AvgDriverQuali'] = df_model['AvgDriverQuali'].fillna(df_model['AvgTeamQuali'])
df_model['AvgTeamQuali'] = df_model['AvgTeamQuali'].fillna(grid_average)

# Feature & target
X = df_model[['AvgDriverQuali', 'AvgTeamQuali']]
y = df_model['Position']

X.head(), y.head()
```

```python
# =========================
# CELL 9  TRAIN & EVALUATE MODEL
# =========================

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Model
rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf.fit(X_train, y_train)

# Evaluation
y_pred = rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE : {mae:.2f}")
print(f"R  : {r2:.2f}")
```

```python
# =========================
# CELL 10  PREDICT GRID 2026
# =========================

X_2026 = grid_2026[['AvgDriverQuali', 'AvgTeamQuali']]

grid_2026['PredictedQuali'] = rf.predict(X_2026)

# Urutkan seperti starting grid
grid_2026 = grid_2026.sort_values('PredictedQuali').reset_index(drop=True)
grid_2026['GridPosition'] = grid_2026.index + 1

grid_2026[['GridPosition', 'DriverCode', 'Team', 'PredictedQuali']]
```

```python
# =========================
# CELL 11  VISUALIZATION
# =========================

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 10))
plt.barh(
    grid_2026['DriverCode'][::-1],
    grid_2026['PredictedQuali'][::-1]
)

plt.xlabel('Predicted Qualifying Position')
plt.title('Predicted Starting Grid  Australian GP 2026')
plt.tight_layout()
plt.show()
```
